```python
ax = sns.countplot(data = df, x = 'artist_count')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(25,8))
ax = sns.countplot(data = df, x = 'released_year')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(25,8))
ax = sns.countplot(data = df, x = 'released_month')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
spot_art = df.groupby(['artist(s)_name'], as_index=False)['artist_count'].sum().sort_values(by = ['artist_count'], ascending=False).head(10)
sns.set(rc = {'figure.figsize' : (16,5)})
sns.barplot(x = 'artist(s)_name', y = 'artist_count', data = spot_art)
```

```python
gb = df.groupby("artist(s)_name").agg({'valence_%':'mean','energy_%':'mean','acousticness_%':'mean','instrumentalness_%':'mean','liveness_%':'mean','speechiness_%':'mean'}).sort_values(by = ['artist(s)_name'], ascending=False).head(5)
sns.heatmap(gb, annot = True)
plt.show()
```

From the above data, we have analyized that: 
- Most songs are released in yrs 2023 and 2021, in the month of january, may and march.
- Maximum songs have single artist.
- J. Cole songs is best in terms of acousticness, energy, liveness etc.
- Taylor Swift and TheWeeknd are the two most popular and maximum songs producers.
